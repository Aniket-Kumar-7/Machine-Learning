```python
col=['Age','RestingBP','Cholesterol','MaxHR']
i=0
for cols in col:
  i=i+1
  plt.subplot(2,2,i)
  sbn.histplot(df[cols],kde=True)

plt.tight_layout()
```

```python
df['Cholesterol'].value_counts()
```

```python
cholesterol_mean=df.loc[df['Cholesterol']!=0,'Cholesterol'].mean()
cholesterol_mean
```

```python
df['Cholesterol']=df['Cholesterol'].replace(0,cholesterol_mean).round(2)
```
