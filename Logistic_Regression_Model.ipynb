```python
df.drop(['deck'],axis=1,inplace=True)
```

```python
df.isnull().sum()/len(df)
```

```python
df['age'].fillna(df['age'].median(),inplace=True)
```

```python
df.head()
```

```python
df.drop(['alive','who'],axis=1,inplace=True)
```

```python
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
df.loc[df['age']>=18,'adult_male']=True
df.head()
```

```python
df.head()
```

```python
df['embarked'].value_counts()[0]/len(df)*100
```

```python
df.drop(['adult_male'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(['parch'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop(['embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['sex']=df['sex'].map({'male':1,'female':0})
```

```python
df.head()
```

```python
df['class'].value_counts()
df['class']=df['class'].map({'Third':3,'First':1,'Second':2})
```

```python
df.head()
df['alone']=df['alone'].map({False:0,True:1})
df.head()
```

```python
df['age']=df['age'].astype(int)
df['fare']=df['fare'].round().astype(int)
```

```python
df.head()
df['embark_town'].value_counts()
```

```python
df['embark_town'].unique()
df['embark_town']=df['embark_town'].map({'Southampton':1,'Cherbourg':2,'Queenstown':3})
```

```python
df.head()
df=df.astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
df['embark_town']=label.fit_transform(df['embark_town'])
df['sex']=label.fit_transform(df['sex'])
```

```python
df['embark_town'].value_counts()
df.head()
```

```python
X=df.drop('survived',axis=1)
y=df['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test).round().astype(int)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
